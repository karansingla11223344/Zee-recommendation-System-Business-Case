```
def top_5_similar_movies(base_movie_1):

  #merge the ratings and movies dataframe
  mov_rat=pd.merge(ratings,movies,on="MovieID")

  #create pivot table
  user_movie_matrix_1=mov_rat.pivot_table(index='UserID',columns='Title',values='Rating')
  user_movie_matrix_1

  #choose a base movie
  base_move_ratings=user_movie_matrix_1[base_movie_1]
  base_move_ratings
  similar_movies=user_movie_matrix_1.corrwith(base_move_ratings)
  corr_df=pd.DataFrame(similar_movies,columns=['PearsonCorr'])
  corr_df.dropna(inplace=True)

  #group by title to found count and mean of rating
  movie_stats=mov_rat.groupby('Title')['Rating'].agg(['count','mean'])
  corr_df=corr_df.join(movie_stats)

  #filtered the rows based on minimum count of ratings is 50
  filtered_corr=corr_df[corr_df['count']>=50]
  top_5_similar=filtered_corr.sort_values('PearsonCorr',ascending=False).head(6)
  return top_5_similar.head(6)
```

```

from sklearn.neighbors import NearestNeighbors
```

```
# Fit KNN on item-item matrix
model_knn = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn.fit(item_matrix)
```

```
# Function to recommend top 5 similar movies given a movie_id
def get_similar_movies(movie_id, movies, n_neighbors=5):

    if movie_id not in item_matrix.index:
        return f"movie_id {movie_id} not found in matrix."

    movie_vector = item_matrix.loc[movie_id].values.reshape(1, -1)
    distances, indices = model_knn.kneighbors(movie_vector, n_neighbors=n_neighbors+1)

    similar_ids = item_matrix.index[indices.flatten()][1:]  # skip the first (itself)

    # Get movie names from movie_id
    return movies[movies['MovieID'].isin(similar_ids)]
```

```
import pandas as pd

new_user_df = pd.DataFrame.from_dict(user_ratings, orient='index', columns=['rating'])
new_user_df
```

```
common_movies = list(user_ratings.keys())
```

```
user_title_df=mov_rat.pivot_table(index='UserID',columns='Title',values='Rating')
```

```


# Filter users who have rated ALL the movies new user has rated
filtered_users = user_title_df.dropna(subset=common_movies, how='all')
```

```
filtered_users['common_count'] = filtered_users[common_movies].notna().sum(axis=1)

# Sort by count and take top 100
top_100_users = filtered_users.sort_values('common_count', ascending=False).head(100)
```

```
from scipy.stats import pearsonr

def get_similarity(row):
    common = new_user_df.index.intersection(row.dropna().index)
    if len(common) < 2:
        return 0  # Not enough data to compute correlation
    return pearsonr(new_user_df.loc[common]['rating'], row[common])[0]

top_100_users['similarity'] = top_100_users.apply(get_similarity, axis=1)
```

```
top_10_users = top_100_users.sort_values('similarity', ascending=False).head(10)

top_10_users
```

```
# Movies rated by these users
top_10_movie_ratings = user_title_df.loc[top_10_users.index]
top_10_movie_ratings
```

```
# Multiply rating by similarity for each rating
weighted_ratings = top_10_movie_ratings.T.dot(top_10_users['similarity'])

# Sum of similarities (per movie)
similarity_sums = (top_10_movie_ratings.notna().T.dot(top_10_users['similarity']))
similarity_sums=similarity_sums.replace(0,np.nan)
# Final score: weighted average
recommendation_scores = weighted_ratings / similarity_sums
```

```
# Drop movies already rated by new user
recommendation_scores = recommendation_scores.drop(labels=common_movies, errors='ignore')

# Get top 5 movie recommendations
top_5_recommendations = recommendation_scores.sort_values(ascending=False).head(5)
print(top_5_recommendations)
```
